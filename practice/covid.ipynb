```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import Image
from IPython.core.display import HTML 
```

```python
df = pd.read_csv('COVID clinical trials.csv')
df
```

```python
df['Status'].value_counts().plot(kind='barh')
```

```python
df.head().T
```

```python
df['Conditions'].value_counts()
```

```python
df['Study Type'].value_counts().plot(kind='barh')
```

```python
interventional = df[df['Study Type']=='Interventional']
top_interventions = interventional['Interventions'].value_counts().sort_values(ascending=True)[-5:]
plt.pie(top_interventions.values, labels=top_interventions.keys())
plt.show()
```

```python
df[['Name', 'City', 'State', 'Country']] = df['Locations'].str.rsplit(', ', n=3, expand=True)
```

```python
interventional[['Name', 'City', 'State', 'Country']] = interventional['Locations'].str.rsplit(', ', n=3, expand=True)
```

```python
plt.figure(figsize=(17,8))
interventional['Country'].value_counts().sort_values(ascending=True)[-20:].plot(kind='bar')
```

```python
plt.figure(figsize=(10,8))
interventional['Phases'].value_counts().plot(kind='pie', autopct='%.2f')
```
